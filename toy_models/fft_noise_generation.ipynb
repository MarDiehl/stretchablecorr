```python
import numpy as np
import matplotlib.pylab as plt
from scipy.fft import ifft2
from scipy.ndimage import gaussian_filter
```

- fractal 2D noise using Power Spectral Density

```python
size = 800
```

```python
k = np.arange(size) + 1
kx, ky = np.meshgrid(k, k)
kr = np.sqrt(kx**2 + ky**2)
```

```python
amplitudes = 1e6*( .1 + 1*kr**1.2)/( 100 + kr**(2.6) )
phases = np.random.rand(size, size)
tf = amplitudes * np.exp(2*np.pi*phases*1j)
image = ifft2(tf)
image = np.abs(image)
```

```python
plt.loglog(amplitudes[:, 0], '-', linewidth=4); plt.grid(which='both', linewidth=1, color='#bbb');
plt.xlabel('1/freq'); plt.ylabel('amplitude');
```

```python
plt.imshow(image); plt.colorbar();
```

```python
plt.hist(image.flatten(), bins=60); plt.xlabel('intensity');
```

```python
second_image = image**0.55
plt.imshow(second_image); plt.colorbar();
```
