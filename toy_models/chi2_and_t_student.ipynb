```python
import numpy as np
import matplotlib.pylab as plt
```

```python
from scipy import stats
#import chi2(*args, **kwds
```

```python
def sample_chi2(N):
    X = np.random.randn(N)
    return np.sum( X**2 )

N = 5
samples = [sample_chi2(N) for _ in range(2000)]

x = np.linspace(0, np.max(samples), 123)
pdf = stats.chi2.pdf(x, N, loc=0, scale=1)

plt.hist(samples, density=True, bins=20)
plt.plot(x, pdf, linewidth=4)
plt.xlabel('chi2');
plt.ylabel('PDF chi2');
```

```python
def sample_student(N):
    #  mu = 0
    X = np.random.randn(N)
    x_mean = np.mean(X)
    S = np.sqrt( np.sum( (X - x_mean)**2  )/( N-1 ) )
    t = x_mean / (S / np.sqrt(N))
    return t

N = 30
samples = [sample_student(N) for _ in range(2000)]

x = np.linspace(-np.max(samples), np.max(samples), 123)
pdf = stats.t.pdf(x, N-1, loc=0, scale=1)

plt.hist(samples, density=True, bins=20)
plt.plot(x, pdf, linewidth=4)
plt.xlabel('t');
plt.ylabel('PDF t (student)');
```

- chi2: sum of X^2 with known variance & mean --> CI for estimated variance of knwon distribution
- Student: mean estimator distribution with unknown variance (i.e. estimated too) --> confidence interval for the true mean
